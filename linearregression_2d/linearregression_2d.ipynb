```python
import numpy as np
```

```python
import matplotlib.pyplot as plt 
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
import pandas as pd
```

```python
#load data
X=[]
Y=[]
for line in open('data_2d.csv'):
    x1,x2,y = line.split(',')
    X.append([float(x1),float(x2),1])
    Y.append(float(y))
```

```python
X=np.array(X)
Y=np.array(Y)
X
```

```python
#plotting the data
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

ax.scatter(X[:,0],X[:,1],Y)
plt.show()
```

```python
#calculation
a=np.dot(X.T,X)
b=np.dot(X.T,Y)
w=np.linalg.solve(a,b)

yeq=np.dot(X,w)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],Y)
ax.plot(sorted(X[:,0]),sorted(X[:,1]),sorted(yeq))
```

```python
#using r square method
d1=Y-yeq
d2=Y-Y.mean()
r_Square=1-(d1.dot(d1)/d2.dot(d2))
```

```python
#As r_square is nearly equal to one hence good model
r_Square
```
